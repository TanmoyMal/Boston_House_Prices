```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

data = pd.read_csv("/content/housing.csv",sep="\s+",names=['CRIM',	'ZN',	'INDUS',	'CHAS',	'NOX',	'RM',	'AGE',	'DIS',	'RAD',	'TAX',	'PTRATIO',	'B',	'LSTAT',	'MEDV'])

data.head(10)
```

```python
# Plot histograms for all columns
data["AGE"].hist()
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
data.plot(kind='scatter', x='AGE', y='TAX')
plt.title("Age vs Tax")
plt.xlabel("AGE")
plt.ylabel("TAX")
plt.show()
```

```python
sns.boxplot(data=data[['CRIM', 'ZN', 'INDUS', 'NOX', 'RM']])
plt.title("Box plot of selected features")
plt.show()
```

```python
sns.displot(data, x="MEDV", kde=True)
plt.title("Distribution plot of MEDV")
plt.show()
```

```python
sns.heatmap(data, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# Pairplot visualize
sns.pairplot(data)
plt.suptitle('Pairplot of Boston House Prices', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='TAX', y='AGE', data=data)
plt.xticks(rotation=90)
plt.title('Average Tax')
plt.show()
```
